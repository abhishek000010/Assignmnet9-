```python
Q2 -> Differentiate between Abstraction and Encapsulation. Explain with an example.

Ans -> Abstraction is a process of hiding the implementation details and exposing only the essential features to the users. In Python,
we can achieve abstraction using abstract classes or interfaces. An abstract class is a class that cannot be instantiated, and its purpose
is to provide a template for its subclasses. Abstract classes can define abstract methods, which must be implemented by the subclasses.

# example of abstractiion
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
class Rectangle(Shape):
    def __init__(self, length , breadth):
        self.length = length 
        self.breadth = breadth 
    def area(self):
        return self.length * self.breadth
    
class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius
        
    def area(self):
        return 3.14 * self.radius * self.radius
r = Rectangle(10 , 5)
print("Area of Rectangle :" , r.area())

c = Circle(7)
print("Area of Circle :", c.area())

# Encapsulation

Encapsulation is a process of hiding the internal details of an object and exposing only the necessary
methods to interact with it. In Python, we can achieve encapsulation using private and protected variables and methods.

class BankAccount:
    def __init__(self , account_no , balance):
        self.__account_no = account_no
        self.__balance = balance
        
    def deposit(self , amount):
        self.__balance += amount
        
    def withdraw(self , amount):
        if(amount > self.__balance):
            print("amount is insufficient")
        else:
            self.__balance -= amount
    def get_balance(self):
        return self.__balance
    
account = BankAccount("3847823723" , 5000)
account.deposit(2000)
account.withdraw(1000)
print("current Balance : " ,account.get_balance())
```

```python
Q3 -> What is abc module in python? Why is it used?

Ans -> The abc module in Python stands for "Abstract Base Classes". It is used to define abstract base classes, which are classes that cannot be instantiated themselves, but can be subclassed to create concrete classes.
Abstract base classes are useful for creating a common interface or API that multiple concrete classes can implement in their own way. For example, if you have a set of classes that need to perform some common functionality,
you can define an abstract base class that defines that functionality, and then have each concrete class subclass the abstract base class to implement that functionality in their own way.
The abc module provides tools for creating and working with abstract base classes in Python. It includes the ABC class, which is the base class for defining abstract base classes, as well as decorators like @abstractmethod and 
@abstractproperty for defining abstract methods and properties that must be implemented by concrete subclasses.
In summary, the abc module is used to create abstract base classes that provide a common interface or API for multiple concrete classes to implement in their own way. This can make your code more modular, reusable, and easier to maintain over time.
```

```python
Q4 -> How can we achieve data abstraction?

Ans -> Data abstraction is a programming concept that refers to the ability to hide the implementation details of data objects from the users of that data. This can be achieved in a number of ways, including:

1) Encapsulation: Encapsulation is the practice of bundling data and methods that operate on that data into a single unit, known as a class. By defining a class, you can control how users of that class interact with its data, and prevent them from accessing or modifying that data directly.

2) Abstract Classes and Interfaces: Abstract classes and interfaces are used to define a common interface or API that multiple concrete classes can implement in their own way. By defining abstract classes or interfaces, you can hide the implementation details of concrete classes from their users, and provide a more generalized view of the data.

3) Access Modifiers: Access modifiers, such as public, private, and protected, can be used to control how data is accessed and modified in a program. By defining data members as private, for example, you can prevent users from directly accessing or modifying that data.

4)Data Hiding: Data hiding is the practice of restricting access to certain data members or methods of a class, in order to prevent users from modifying or accessing them directly. This can be achieved by defining data members as private, and providing access to them only through getter and setter methods.

In summary, data abstraction can be achieved through techniques such as encapsulation, abstract classes and interfaces, access modifiers, and data hiding. By using these techniques, you can hide the implementation details of data objects from their users, and provide a more abstract and generalized view of the data. This can make your code more modular, reusable, and easier to maintain over time.
```
